```python
import glob
from pathlib import Path
import re
import os
```

```python
fit_files = glob.glob("./fits/*.fits")
```

```python
for file in fit_files:
    file_path = Path(file)
    run_num = re.findall("([0-9]+)", file_path.name)[0]
    if not(os.path.exists(run_num)):
        os.system(f"mkdir {run_num}")
    base_dir = f"./{run_num}/"
    config_file = base_dir+"config.yaml"
    print(run_num)
    
    vtspy.JointConfig(file, config_file=config_file, 
                      fermi_outdir = base_dir+"/fermi/", 
                      fermi_data = base_dir+"/fermi/",
                      joint_outdir = base_dir+"/joint/",
                      veritas_outdir = base_dir+"/veritas/")
    veritas = vtspy.VeritasAnalysis(config_file=config_file)
    veritas.construct_dataset(eff_cut=10, bias_cut=10)
    veritas.fit(model="LogParabola")
    logP_fit = veritas.fit_results
    veritas.fit(model="PowerLaw")
    pl_fit = veritas.fit_results
    veritas.analysis()
    veritas.print_flux()
    break
```
